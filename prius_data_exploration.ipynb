# Prius Data Exploration
### Data Setup

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = "output12-21-15-39-32-filtered.csv"
unfiltered = "output12-21-15-39-32.csv"

data = np.genfromtxt(unfiltered, delimiter=',')
```

```python
# first 3 columns are incompatable with numpy, working on this.
# Using indexes for x-axis for now, Goal is to use time someday....
data = data[10:, 3:] # for now cut out
```

```python
brake_pedal = data[:, 3]
gas_pedal = data[:, 4]
em_amps = data[:, 5]
batt_volts = data[:, 6]
rpm = data[:, 7]
coolant_temp = data[:, 8]
steering_angle = data[:, 9]
gps_speed = data[:, 13]
car_speed = data[:, 2]
```

```python
x = np.arange(0, data.shape[0], 1) # Generate x axis
```

# Plots
### Car Speed vs GPS Speed

```python
plt.figure(figsize=(12,8), dpi= 100)
plt.plot(x, car_speed, c='green', label='Car Speed', alpha=0.6)
plt.plot(x, gps_speed, c='red', label='GPS Speed', alpha=0.6)
plt.ylabel('Speed (MPH)')
plt.legend()
plt.show()
```

### Pedals and Amp Data

```python
plt.figure(figsize=(12,8), dpi= 100)
plt.plot(x, gas_pedal, c='green', label='Gas Pedal %', alpha=0.6)
plt.plot(x, brake_pedal, c='red', label='Brake Pedal %', alpha=0.6)
plt.plot(x, em_amps, c='yellow', label='EM Amps', alpha=0.6)
plt.legend()
plt.show()
```

### Pedals and Volt Data

```python
plt.figure(figsize=(12,8), dpi= 100)
plt.plot(x, gas_pedal, c='green', label='Gas Pedal %', alpha=0.6)
#plt.plot(x, brake_pedal, c='red', label='Brake Pedal %', alpha=0.6)
plt.plot(x, batt_volts - 201, c='orange', label='Battery Voltage Offset', alpha=0.6) # Normalizes voltage
plt.legend()
plt.show()
```

### Engine RPMs and Temp

```python
plt.figure(figsize=(12,8), dpi= 100)
plt.plot(x, rpm / 10, c='brown', label='RPMs', alpha=0.6)
plt.plot(x, coolant_temp, c='red', label='Coolant Temp (F)', alpha=0.6)
plt.legend()
plt.show()
```

### Steering Angle and Pedals

```python
plt.figure(figsize=(12,8), dpi= 100)
plt.plot(x, steering_angle, c='green', label='Steering Angle ()', alpha=0.6)
plt.plot(x, brake_pedal, c='red', label='Brake Pedal %', alpha=0.6)
plt.legend()
plt.show()
```
